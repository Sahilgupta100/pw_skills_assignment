Question:
Write a Python program to calculate the mean, median, and mode of a given dataset.

```python
import statistics

# Sample dataset
data = [10, 20, 20, 30, 40, 50, 50, 50, 60, 70]

# Calculating mean, median, and mode
mean_value = statistics.mean(data)
median_value = statistics.median(data)
mode_value = statistics.mode(data)

# Displaying results
print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
print(f"Mode: {mode_value}")
```

Write a Python program to compute the variance and standard deviation of a given dataset.

```python
import statistics

# Sample dataset
data = [10, 20, 30, 40, 50, 60, 70]

# Calculating variance and standard deviation
variance_value = statistics.variance(data)
std_dev_value = statistics.stdev(data)

# Displaying results
print(f"Variance: {variance_value}")
print(f"Standard Deviation: {std_dev_value}")
```

Question:
Write a Python program to implement random sampling and stratified sampling.

```python
import random
import numpy as np
from sklearn.model_selection import train_test_split

# Sample dataset
population = list(range(1, 101))  # Numbers from 1 to 100

# Random Sampling: Selecting 10 random elements
random_sample = random.sample(population, 10)
print(f"Random Sample: {random_sample}")

# Stratified Sampling (Using categories)
labels = np.random.choice(["A", "B", "C"], size=100)  # Assign random categories
train_data, _, train_labels, _ = train_test_split(population, labels, stratify=labels, test_size=0.8)

print(f"Stratified Sample: {train_data}")
```

Write a Python function to calculate the range of a dataset.

```python
def calculate_range(data):
    return max(data) - min(data)

# Sample dataset
data = [15, 20, 35, 40, 50, 65]

# Calculate range
range_value = calculate_range(data)
print(f"Range of dataset: {range_value}")
```

Create a dataset and plot its histogram to visualize skewness.

```python
import matplotlib.pyplot as plt
import numpy as np

# Creating a skewed dataset (positively skewed)
data = np.random.exponential(scale=2, size=1000)

# Plot histogram
plt.hist(data, bins=30, color='blue', edgecolor='black', alpha=0.7)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram to Visualize Skewness")
plt.show()
```

Question:
Calculate the skewness and kurtosis of a dataset using Python libraries.

```python
from scipy.stats import skew, kurtosis

# Sample dataset
data = [10, 20, 30, 40, 50, 60, 70, 100, 200, 300]

# Calculate skewness and kurtosis
skewness = skew(data)
kurt = kurtosis(data)

print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurt}")
```

Question:
Generate a dataset and demonstrate positive and negative skewness.

```python
import seaborn as sns

# Generating skewed datasets
positive_skew = np.random.exponential(scale=2, size=1000)
negative_skew = -np.random.exponential(scale=2, size=1000)

# Plot histograms
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.histplot(positive_skew, bins=30, kde=True, ax=axes[0])
axes[0].set_title("Positive Skewness")

sns.histplot(negative_skew, bins=30, kde=True, ax=axes[1])
axes[1].set_title("Negative Skewness")

plt.show()
```

Question:
Write a Python script to calculate covariance between two datasets.

```python
import numpy as np

# Sample datasets
X = [10, 20, 30, 40, 50]
Y = [15, 25, 35, 45, 55]

# Calculate covariance
cov_matrix = np.cov(X, Y)
print(f"Covariance: {cov_matrix[0][1]}")
```

Question:
Write a Python script to calculate the correlation coefficient between two datasets.

```python
# Calculate correlation coefficient
correlation_coefficient = np.corrcoef(X, Y)[0][1]
print(f"Correlation Coefficient: {correlation_coefficient}")
```
