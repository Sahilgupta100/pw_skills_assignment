##Question 5. Train a Decision Tree Classifier and visualize the tree using graphviz


```python
from sklearn.tree import export_graphviz
import graphviz
from sklearn.datasets import load_iris
iris = load_iris()
clf = DecisionTreeClassifier()
clf.fit(iris.data, iris.target)

export_graphviz(clf, out_file="tree.dot", feature_names=iris.feature_names, class_names=iris.target_names,
                filled=True, rounded=True, special_characters=True)

with open("tree.dot") as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```

##Question 6. Train a Decision Tree Classifier with max depth 3 and compare accuracy with full tree

```python
clf_limited = DecisionTreeClassifier(max_depth=3)
clf_limited.fit(X_train, Y_train)

print("Accuracy with max_depth=3:", accuracy_score(Y_test, clf_limited.predict(X_test)))
print("Accuracy with full tree:", accuracy_score(Y_test, clf.predict(X_test)))
```

##Question 7. Train a Decision Tree Classifier with min_samples_split=5 and compare accuracy


```python
clf_split = DecisionTreeClassifier(min_samples_split=5)
clf_split.fit(X_train, Y_train)

print("Accuracy with min_samples_split=5:", accuracy_score(Y_test, clf_split.predict(X_test)))
print("Accuracy with default tree:", accuracy_score(Y_test, clf.predict(X_test)))
```

##Question 8. Apply feature scaling before training and compare accuracy


```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(iris.data)

X_train_scaled, X_test_scaled, y_train, y_test = train_test_split(X_scaled, iris.target, test_size=0.2, random_state=42)

clf_scaled = DecisionTreeClassifier()
clf_scaled.fit(X_train_scaled, y_train)

print("Accuracy with scaled data:", accuracy_score(y_test, clf_scaled.predict(X_test_scaled)))
```

##Question 9. Train a Decision Tree Classifier using One-vs-Rest strategy

##Question 11. Train a Decision Tree Regressor with max_depth=5 and compare

```python
reg = DecisionTreeRegressor()
reg.fit(X_train, Y_train)


reg_limited = DecisionTreeRegressor(max_depth=5)
reg_limited.fit(X_train, y_train)

print("MSE with max_depth=5:", mean_squared_error(Y_test, reg_limited.predict(X_test)))
print("MSE with full tree:", mean_squared_error(Y_test, reg.predict(X_test)))
```

##Question 12. Apply Cost Complexity Pruning (CCP) and visualize effect


```python
path = clf.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas = path.ccp_alphas

accuracies = []
for ccp_alpha in ccp_alphas:
    pruned_clf = DecisionTreeClassifier(ccp_alpha=ccp_alpha)
    pruned_clf.fit(X_train, y_train)
    acc = accuracy_score(y_test, pruned_clf.predict(X_test))
    accuracies.append(acc)

plt.plot(ccp_alphas, accuracies, marker='o')
plt.xlabel("ccp_alpha")
plt.ylabel("Accuracy")
plt.title("Effect of CCP on Accuracy")
plt.show()
```
